```python
with open("sureler.txt") as sureler:
    sureler = sureler.read().splitlines()
sureler 
```

```python
import json

kuran = {}
for sure in sureler:
    kuran[sure] = {}

with open('kuran.json', 'w') as f:
    f.write(json.dumps(kuran, ensure_ascii=False))
```

```python
import requests
import re
from bs4 import BeautifulSoup

for n, sure in enumerate(kuran):
    print(sure, n+1, "/", len(kuran))
    r = requests.get("https://tevhidmeali.com/sure/" + sure)

    source = BeautifulSoup(r.content,"lxml")

    ayetler = []
    for ayet in source.select(".row.ayetler"):
        ayetler.append(re.sub(' +', ' ', ayet.text.strip()))

    kuran[sure] = ayetler

with open('kuran.json', 'w') as f:
    f.write(json.dumps(kuran, ensure_ascii=False))
```
